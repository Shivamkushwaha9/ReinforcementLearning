### TRUST REGION
#### Trust region methods are optimization techniques that improve a solution (like a policy in reinforcement learning) incrementally by constraining each update step to stay within a "trusted region" where the approximations used in optimization remain valid.

*TRUST REGION IS NOTHING BUT approximating solution to constrain it in specific range/region.*
```
Core idea: When updating a policy, don't change it too much at once. Large changes can lead to performance collapse.
```

`
This region is typically defined using a distance metric like KL divergence between the old and new policies
`
