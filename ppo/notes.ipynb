# **4. Generalized Advantage estimator**
### **GAE's a generalization that includes the other three as special cases. It's named "generalized" precisely because it provides a single formula that can represent any of the other advantage estimators based on the λ parameter.**


### When λ=0, GAE equals the one-step TD error. 
### When λ=1, GAE equals the full advantage.
### Values between 0 and 1 give you a weighted combination of different n-step advantages
